```python
import os
import pandas as pd
import geopandas as gpd

from shapely.geometry import Point
import fiona
```

```python
network_file = 'Chatt_Master.shp'
network_path = os.path.join('..', 'Data' )
```

```python
network = gpd.read_file(os.path.join(network_path, network_file))
network = network[['ID', 'geometry']]

network['A'] = [pt['geometry'].coords[0] for i,pt in network.iterrows()]
network['B'] = [pt['geometry'].coords[-1] for i,pt in network.iterrows()]

# network = network.drop(columns='geometry')

# print(network.crs)
# print('Check for duplicates:', network['ID'].duplicated().any())
# network['ID'].describe()
```

### NODES

```python
nodes = gpd.read_file(os.path.join(network_path, 'Chatt_Master_TransitStop.shp'))
nodes = nodes[['ID', 'geometry']]
nodes['node_coords'] = [(c.x, c.y) for c in nodes['geometry']]

nodes.plot()
nodes.head()
```

```python
network_stream = []

for i, pt in nodes.iterrows():
    net = network[network['geometry'].intersects(pt['geometry'])]
    for j, stream in net.iterrows():
        if pt['node_coords']==stream['A']:
            print(pt['ID'], 'A', stream['ID'])
            network_stream.append([pt['ID'], 'A', stream['ID']])
        elif pt['node_coords']==stream['B']:
            print(pt['ID'], 'B', stream['ID'])
            network_stream.append([pt['ID'], 'B', stream['ID']])

network_stream = pd.DataFrame(network_stream, columns=['id', 'end', 'link_id'])
```

```python
network_stream.to_csv(os.path.join(network_path, 'network_nodes.csv'))
```
